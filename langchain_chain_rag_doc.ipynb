```python
# configuration
docs_dir_path="C:\\Users\\Sachin Maurya\\OneDrive\\Desktop\\projects\\Rag_Project\\docs"
vector_db_path="C:\\Users\\Sachin Maurya\\OneDrive\\Desktop\\projects\\Rag_Project\\vector_db"
collection_name="document_collection"
```

```python
# loading the embedding model
embedding=HuggingFaceEmbeddings()
```

```python
loader = DirectoryLoader(
    path = docs_dir_path,
    glob="./*.pdf",
    loader_cls=UnstructuredFileLoader 
)
```

```python
# load the document
documents= loader.load()
```

```python
text_spliter=CharacterTextSplitter(
    chunk_size=2000,
    chunk_overlap=500
 
)
```

```python
# splitting the text into smaller chunks
text_chunks=text_spliter.split_documents(documents)
```

```python
# creating the vector store
vector_store=Chroma.from_documents(
    documents=text_chunks,
    embedding=embedding,
    persist_directory=vector_db_path,
    collection_name=collection_name
)
```
